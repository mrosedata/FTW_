## SQL

The Northwind sample database was provided with Microsoft Access as a tutorial schema for managing small business customers, orders, inventory, purchasing, suppliers, shipping, and employees. Northwind is an excellent tutorial schema for a small-business ERP, with customers, orders, inventory, purchasing, suppliers, shipping, employees, and single-entry accounting.

Sqlite (https://www.sqlitetutorial.net/) and SQLAlchemy (https://www.geeksforgeeks.org/sqlalchemy-introduction/)

```python
import sqlite3

conn = sqlite3.connect("../data/northwind.db") # ensure this is the correct path for the sqlite file.


northwind_sqlite_df = pd.read_sql_query("select * from Customers limit 10", conn)


display(northwind_sqlite_df.head())

conn.close()
```

```python
# Example

# QUESTION: Compute the total qty product sold per product sorted by highest to lowest qty

# SQL version
conn = sqlite3.connect("../data/northwind.db") # ensure this is the correct path for the sqlite file.
query_df = pd.read_sql_query("select ProductID,sum(Quantity) as qty_sold from 'Order Details' group by ProductID order by qty_sold desc", conn)

display(query_df)
conn.close()
```

```python
# Pandas version
conn = sqlalchemy.create_engine('sqlite:///../data/northwind.db') # ensure this is the correct path for the sqlite file.
query_df = pd.read_sql_table("Order Details",conn) # retrieve the whole table and store in df
query_df = query_df[['ProductID','Quantity']].groupby('ProductID').sum().sort_values('Quantity',ascending=False)
query_df = query_df.reset_index(drop=False)
display(query_df)
```
